```python
import os 
import re
import math
import numpy as np
import ast
from functools import cmp_to_key

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input'

def read_txt(filename):
    with open(filename) as data_file:
            
        lines = data_file.read().split('\n\n') 
        lines = [line.split('\n') for line in lines]
        
    return lines

def compare(left, right):
    """
    -1: False 1: True 0: Continue
    """
    if isinstance(left, int) and isinstance(right, int):
        return compareint(left, right)
    else:
        left = [left] if isinstance(left, int) else left
        right = [right] if isinstance(right, int) else right
        
        return comparelist(left, right)
    
def compareint(left, right):
    if left < right:
        return 1
    elif left == right:
        return 0
    else:
        return -1
    
    
def comparelist(left, right):
        
    if len(left) == 0 and len(right) != 0:
        return 1
    elif len(left) != 0 and len(right) == 0:
        return -1
    else:
        if left == right:
            return 0
        if len(left) <= len(right):
            min_l = len(left)
            for i in range(min_l):
                res = compare(left[i], right[i])
                if res != 0:
                    return res
        
            return 1
        
        else:
            min_l = len(right)
            for i in range(min_l):
                res = compare(left[i], right[i])
                if res != 0:
                    return res
                
            return -1
    
```

```python
l = read_txt(PATH+'/day13.txt')
pairs = []

s = 0

for idx, pair in enumerate(l):
    left = ast.literal_eval(pair[0])
    right = ast.literal_eval(pair[1])
    if compare(left, right) == 1:
        s += idx+1
    pairs += [left, right]
    

print(f'part 1: {s}') # part 1

pairs += [[[2]], [[6]]]
pairs2 = sorted(pairs, key = cmp_to_key(compare))
pairs2.reverse()
a = pairs2.index([[2]]) + 1
b = pairs2.index([[6]]) + 1

print(f'part 2: {a * b}') # part 2
```
