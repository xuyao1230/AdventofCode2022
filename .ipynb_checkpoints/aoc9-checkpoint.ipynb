```python
import os
import numpy as np

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input/'

def read_txt(filename):
    with open(filename) as file:
        lines = file.readlines()
        lines = [line.rstrip().split(" ") for line in lines]

    return lines

def move(head, action):
    x = head[0]
    y = head[1]
    
    if action == 'U':
        x = x - 1
    elif action == 'D':
        x = x + 1
    elif action == 'L':
        y = y - 1
    else:
        y = y + 1
        
    return (x,y)

def check(head, tail):
    adjacent = False
    if abs(head[0] - tail[0]) == 0 and abs(head[1] - tail[1]) <= 1:
        adjacent = True
    elif abs(head[1] - tail[1]) == 0 and abs(head[0] - tail[0]) <= 1:
        adjacent = True
    elif abs(head[1] - tail[1]) == 1 and abs(head[0] - tail[0]) == 1:
        adjacent = True
    else:
        adjacent
    return adjacent


def perform(head, tail, motion, visits):
    
    for i in range(0, int(motion[1])):
        h_x = head[0]
        h_y = head[1]
        head = move(head, motion[0])
        if not check(head, tail):
            tail = (h_x, h_y)
            visits.append(tail)
            
    return head, tail, visits

def perform2(motion, visits, knots):
    
    for i in range(0, int(motion[1])):
        
        knots[0] = move(knots[0], motion[0])
        
        for k in range(1, len(knots)):
            if not check(knots[k-1], knots[k]):
                diff_x = knots[k-1][0] - knots[k][0]
                diff_y = knots[k-1][1] - knots[k][1]
                move_x = 0
                move_y = 0
                if diff_x == 0 and diff_y != 0:
                    move_y = 1 if diff_y > 0 else -1
                elif diff_x != 0 and diff_y == 0:
                    move_x = 1 if diff_x > 0 else -1
                else:
                    move_y = 1 if diff_y > 0 else -1
                    move_x = 1 if diff_x > 0 else -1
                    
                knots[k] = (knots[k][0]+move_x, knots[k][1] + move_y)
                if k == len(knots)-1:
                    visits.append(knots[k])
            
    return visits, knots
```

```python
motions = read_txt(PATH+'day9.txt')
############################################ PART 1 ############################################
h = (0,0)
t = (0,0)

visit = [(0,0)]

for m in motions:
    h, t, visit = perform(h,t,m,visit)
print(len(set(visit)))

############################################ PART 2 ############################################
knots = [(0,0)]* 10
visit = [(0,0)]

for m in motions:
    visit, knots = perform2(m, visit, knots)

print(len(set(visit)))
```
