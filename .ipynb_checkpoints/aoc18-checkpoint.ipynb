```python
import os 
import re
import math
import numpy as np
import ast
from functools import cmp_to_key
from collections import defaultdict, deque

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input'


def read_txt(filename):
    with open(filename) as data_file:
        res = []
        lines = data_file.read().split('\n') 
        for line in lines:
            l = line.split(',')
            res.append((int(l[0]), int(l[1]), int(l[2])))
        
        return res


def get_connected(x,y,z):
    return [(x-1, y, z), (x+1, y, z), (x, y-1, z), (x, y+1, z), (x, y, z-1), (x, y, z+1)]


def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))


def out_of_range(x, y, z, minx, maxx, miny, maxy, minz, maxz):
    if minx-1 <= x <= maxx+1 and miny-1 <= y <= maxy+1 and minz-1 <= z <= maxz+1:
        return False
    else:
        return True

def bfs(cubes, minx, maxx, miny, maxy, minz, maxz):
    res = 0
    queue = [(0, 0, 0)]
    visited = set()
    
    while queue:
        x, y, z = queue.pop(0)
        visited.add((x, y, z))
        connected = get_connected(x, y, z)
        for w in connected:
            if w not in visited:
                if w not in queue and not out_of_range(w[0], w[1], w[2], minx, maxx, miny, maxy, minz, maxz):
                    if w in cubes:
                        res += 1
                    else: 
                        queue.append(w)
        
    return res
```

```python
cubes = read_txt(PATH+'/day18.txt')
minx, maxx = min([c[0] for c in cubes]), max([c[0] for c in cubes])
miny, maxy = min([c[1] for c in cubes]), max([c[1] for c in cubes])
minz, maxz = min([c[2] for c in cubes]), max([c[2] for c in cubes])

############################################ PART 1 ############################################
processed = []

res = 0
for c in cubes:
    x = c[0]
    y = c[1]
    z = c[2]
    connected = get_connected(x, y, z)
    ins = intersection(connected, processed)
    sides = len(ins)
    res += 6 - 2*sides
    processed.append(c)


print(f"ans1: {res}")

############################################ PART 2 ############################################
ans2 = bfs(cubes, minx, maxx, miny, maxy, minz, maxz)
print(f"ans2: {ans2}")
```
