```python
import os
from collections import defaultdict

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input/'

def read_txt(filename):
    coordinates = []
    with open(filename) as file:
        lines = file.readlines()
        lines = [line.rstrip() for line in lines]
        for l in lines:
            points = l.split(' -> ')
            coordinates.append(points)
        for i1, l in enumerate(coordinates):
            for i2, p in enumerate(l):
                c = p.split(',')
                coordinates[i1][i2] = (int(c[0]), int(c[1]))
    return coordinates 

def checkrest(l, point):
    down = (point[0], point[1]+1 )
    leftdown = (point[0]-1, point[1]+1)
    rightdown = (point[0]+1, point[1]+1)
    if down not in l:
        return down
    elif leftdown not in l:
        return leftdown
    elif rightdown not in l:
        return rightdown
    else:
        return 1
    
def solution1(rocks):
    h_line = [p[1] for p in rocks]
    bottomy = max(h_line) 
    sand = 0

    stop = False

    while not stop:
        rest = False
        start = (500, 0)
        
        while not rest:
            nex = checkrest(rocks, start)
            if nex == 1:
                rest = True
                rocks[start] = 1
                sand += 1
            else:
                if nex[1] == bottomy:
                    rest = True
                    stop = True
                    
                else:
                    start = (nex[0], nex[1])
    return sand
    

def solution2(rocks):
    h_line = [p[1] for p in rocks]
    bottomy = max(h_line) + 2

    sand = 0
    stop = False

    while not stop:
        rest = False
        start = (500, 0)
        
        while not rest:
            nex = checkrest(rocks, start)
            if nex == 1 or nex[1] == bottomy:
                rest = True
                rocks[start] = 1
                sand += 1
                if start == (500, 0):
                    stop = True
            else:
                start = (nex[0], nex[1])
    

    return sand
    
```

```python
coordinates = read_txt(PATH+'day14.txt')

# I use dict to store the rocks, this will make search much faster than a list
# if use a list, running time for part 2 was about 25 mins

rock = {}
rock2 = {}
for l in coordinates:
    for i in range(0, len(l)-1):
        p1 = l[i]
        p2 = l[i+1]
        if p1[0] == p2[0]:
            for y in range(min(p1[1], p2[1]), max(p1[1], p2[1])+1):
                rock[(p1[0], y)] = 1
                rock2[(p1[0], y)] = 1
        if p1[1] == p2[1]:
            for x in range(min(p1[0], p2[0]), max(p1[0], p2[0])+1):
                rock[(x, p1[1])] = 1
                rock2[(x, p1[1])] = 1


print(solution1(rock))
print(solution2(rock2))
```
