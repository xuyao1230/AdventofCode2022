```python
from collections import deque, defaultdict

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input/'

def read_txt(filename):
    res = {}
    
    with open(filename) as data_file:
            
        lines = data_file.read().split('\n') 
        for i, line in enumerate(lines):
            for j, c in enumerate(line):
                if c == '#':
                    res[(i, j)] = 1
            
    return res

def rotate_dq(dq):
    first = dq.popleft()
    dq.append(first)
    return dq

def get_adjacent(x, y):
    res = {
        'N': (x-1, y),
        'NW': (x-1, y-1),
        'NE': (x-1, y+1),
        'W': (x, y-1),
        'E': (x, y+1),
        'SW': (x+1, y-1),
        'S': (x+1, y),
        'SE': (x+1, y+1)
    }
    return res

def check_one_direc(x, y, ori, dic_ori, eleves, adjacents):
    
    for d in dic_ori[ori]:
        if adjacents[d] in eleves:
            return False
    return True

def check_around(x, y, ori, dic_ori, eleves):
    move = []
    adjacents = get_adjacent(x,y)
    for o in ori:
        if check_one_direc(x, y, o, dic_ori, eleves, adjacents):
            move.append(o)
    if len(move) == 4 or len(move) == 0:
        return 0
    else:
        action = move.pop(0)
        return adjacents[action]
    
def rounds(eleves, ori, dic_ori):
    proposal = defaultdict(list)
    for e in eleves:
        nxt = check_around(e[0], e[1], ori, dic_ori, eleves)
        if nxt == 0:
            proposal[e].append(e)
        else:
            proposal[nxt].append(e)
            
    for p in proposal:
        if len(proposal[p]) == 1:
            eleves.pop(proposal[p][0])
            eleves[p] = 1
      
    same = 0
    for k in proposal:
        if len(proposal[k]) == 1 and proposal[k][0] == k:
            same += 1
    finish = len(proposal) - same
    ori = rotate_dq(ori)
    return eleves, ori, finish

def get_rect(dic):
    ys = [y for (x,y) in dic]
    xs = [x for (x,y) in dic]
    size = (max(ys) - min(ys) + 1) * (max(xs) - min(xs) + 1) - len(dic)
    return size
    
```

```python
eleves = read_txt(PATH+'/day23.txt')
eleves2 = read_txt(PATH+'/day23.txt')

orient = deque(['N','S','W', 'E'])
orient2 = deque(['N','S','W', 'E'])

directions = {
    'N': ['N', 'NW', 'NE'],
    'S': ['S', 'SW', 'SE'],
    'W': ['W', 'NW', 'SW'],
    'E': ['E', 'NE', 'SE']
}

################################################# PART 1 #################################################
r = 0

while r< 10:
    eleves, orient, finish = rounds(eleves, orient, directions)
    r += 1

res = get_rect(eleves)
print(f"ans1: {res}")

################################################# PART 2 #################################################P
i = 0

while True:
    eleves2, orient2, finish = rounds(eleves2, orient2, directions)
    i += 1
    if not finish:
        print(f"ans2: {i}")
        break
```
