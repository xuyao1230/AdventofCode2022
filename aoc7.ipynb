```python
import os 
import re
import numpy as np

PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input'

class Tree(object):
    def __init__(self, name='root', children=None, parent=None, size = 0):
        self.name = name
        self.children = []
        self.parent = parent
        self.size = size
        if children is not None:
            for child in children:
                self.add_child(child)
    
    def add_child(self, node):
        assert isinstance(node, Tree)
        self.children.append(node)
        node.parent = self
        
    def cal_size(self):
        """
        Traverse the tree to calculate the size of each node
        """
        for child in self.children:
            if isinstance(child.name, int):
                self.size += child.size
            else:
                child.cal_size()
                self.size += child.size 
                
    def print_tree(self, level=0):
        """
        https://simonhessner.de/python-3-recursively-print-structured-tree-including-hierarchy-markers-using-depth-first-search/

        """
        print("  " * level, self.name)
    
        for child in self.children:
            child.print_tree(level + 1)
        
    def get_size(self, res = []):
        """
        This method returns a list containing the size of all directories
        """
        if len(self.children) == 0:
            return res
    
        res.append(self.size)
        for child in self.children:
            res = child.get_size(res)
    
        return res



def read_txt(filename):
    with open(filename) as file:
        lines = file.readlines()
        lines = [line.rstrip() for line in lines]

    return lines



def interpret(s, tree):
    """
    This function builds a tree, the name of each node corresponds to a directory name or file size (ugly I know...)
    """
    if s == "$ ls":
        return tree
    elif s.startswith('dir'):
        d = s[4:]
        tree.add_child(Tree(d, []))
    elif s.startswith('$ cd'):
        cd = s[5:]
        if cd != "..":
            for child in tree.children:
                if child.name == cd:
                    tree = child
        else:
            tree = tree.parent
    else:
        files = s.split(' ')
        tree.add_child(Tree(int(files[0]), [], None, int(files[0])))
    
    return tree

```

```python
directories = read_txt(PATH+"/day7.txt")
t = Tree('/', [])

for i in directories:
    t = interpret(i, t)

while t.name != '/':
    t = t.parent
    
#t.print_tree()
t.cal_size()

res = []
res = t.get_size(res)

print(sum([i if i < 100000 else 0 for i in res])) # part 1

enough = list(filter(lambda age: age >= 30000000 - (70000000-t.size), res))
enough.sort()
print(enough[0]) # part 2
```
