```python
import sys
#sys.setrecursionlimit(100000)


PATH = '/Users/xuyaozhang/Desktop/AdventofCode2022/input'


def read_txt(filename):
    with open(filename) as data_file:
            
        lines = data_file.read().split('\n') 
        lines = [int(line) for line in lines]
        
        return lines
    
    
def rotate(l, o, idx):
    """
    Naive, time-consuming solution, operates on a list, move one step in each iteration
    DO NOT use this, only works for small test file
    """
    cur = o[idx] 
    num = l[cur] 
    pos = abs(num)
    for i in range(pos):
        cur = o[idx]
        
        if num > 0:
            if cur == len(l)-1:
                l = [l[0]] + [l[cur]] + l[1:cur]
                o[idx] = 1
                for j in range(len(o)):
                    if j != idx and o[j] != 0:
                        o[j] += 1
            elif cur == len(l)-2:
                l = [l[cur]] + l[0:cur] + [l[-1]]
                o[idx] = 0
                for j in range(len(o)):
                    if j != idx and o[j] != len(o)-1:
                        o[j] += 1
            
            else:
                l[cur] = l[cur+1]
                l[cur+1] = num
                o[idx] += 1
                for j in range(len(o)):
                    if j != idx and o[j] == o[idx]:
                        o[j] -= 1
                
        else:
            if cur == 0:
                l = l[1:-2] + [l[cur]] + [l[-1]]
                o[idx] = len(l) - 2
                for j in range(len(o)):
                    if j != idx and o[j] != len(o)-1:
                        o[j] -= 1
            elif cur == 1:
                l = [l[0]] + l[2:] + [l[cur]]
                o[idx] = len(l) -1
                for j in range(len(o)):
                    if j != idx and o[j] != 0:
                        o[j] -= 1
            else:
                l[cur] = l[cur-1]
                l[cur-1] = num
                o[idx] -= 1
                for j in range(len(o)):
                    if j != idx and o[j] == o[idx]:
                        o[j] += 1
                
    return l, o


def rotate2(l, idx):
    for i in range(len(l)):
        if l[i][0] == idx:
            num = l.pop(i)
            pos = (i + num[1])%len(l)
            # take care of the position in special cases
            if pos == 0 and num[1] != 0:
                pos = len(l) if num[1] <0 else 0
            
            l.insert(pos, (idx, num[1]))
            break
    return l
    
```

```python
nums = read_txt(PATH+'/day20.txt')
len_num = len(nums)
############################################ PART 1 ############################################
l_and_o = [(i, nums[i]) for i in range(len_num)]

for i in range(len_num):
    l_and_o = rotate2(l_and_o, i)

res = [r[1] for r in l_and_o]
idx = res.index(0) 

th1000 = res[(idx+1000)%len_num]
th2000 = res[(idx+2000)%len_num]
th3000 = res[(idx+3000)%len_num]
print(th1000+th2000+th3000)

############################################ PART 2 ############################################
l_and_o_2 = [(i, nums[i]*811589153) for i in range(len_num)]
for r in range(10):
    for i in range(len_num):
        l_and_o_2 = rotate2(l_and_o_2, i)
           
res2 = [r[1] for r in l_and_o_2]
idx2 = res2.index(0) 

th1000 = res2[(idx2+1000)%len_num]
th2000 = res2[(idx2+2000)%len_num]
th3000 = res2[(idx2+3000)%len_num]
print(th1000+th2000+th3000)
```
